```python
import math as m
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# The sigmoid transfer function and its derivatives

```python
def sigma(z):
    return 1 / (1 + m.exp(-z))

def dsigma_dz(z):
    return m.exp(-z) / (1 + m.exp(-z))**2

def d2sigma_dz2(z):
    return (
        2 * m.exp(-2 * z) / (1 + m.exp(-z))**3 - m.exp(-z) / (1 + m.exp(-z))**2
    )

def d3sigma_dz3(z):
    return (
        6 * m.exp(-3 * z) / (1 + m.exp(-z))**4
        - 6 * m.exp(-2 * z) / (1 + m.exp(-z))**3
        + m.exp(-z) / (1 + m.exp(-z))**2
    )
```

```python
z = np.arange(-5, 5, 0.01)
n = len(z)
s = np.zeros(n)
ds_dz = np.zeros(n)
d2s_dz2 = np.zeros(n)
d3s_dz3 = np.zeros(n)
```

```python
for i in range(n):
    s[i] = sigma(z[i])
    ds_dz[i] = dsigma_dz(z[i])
    d2s_dz2[i] = d2sigma_dz2(z[i])
    d3s_dz3[i] = d3sigma_dz3(z[i])
```

```python
plt.plot(z,s,label = "$\sigma(z)$")
plt.plot(z,ds_dz,label = "$d\sigma/dz$")
plt.plot(z,d2s_dz2, label = "$d^2\sigma/dz^2$")
plt.plot(z,d3s_dz3, label = "$d^3\sigma/dz^3$");
plt.xlabel("z")
plt.ylabel("s(z) or derivative")
plt.title("The $\sigma$-function and its first three derivatives")
plt.legend();
```
